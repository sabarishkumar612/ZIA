```python
ds1.describe()
```

From the dataset it is clear that
    - There are no null values in the dataset
    - In most of the columns mean value is almost equal to median. So there is less chance for skewness
    - There are huge difference between 75% and maximum value and hence there is a high possibility for the presence of outliers

```python
#CHECKING FOR OUTLIERS
ds1.plot(kind='box',subplots=True,figsize=(10,10),layout=(6,4))
```

```python
#CHECKING FOR OUTLIERS
from scipy.stats import zscore
f=ds1[['Present_Tmax','Present_Tmin','LDAPS_RHmax','LDAPS_Tmax_lapse','LDAPS_Tmin_lapse','LDAPS_WS','LDAPS_LH','LDAPS_PPT1','LDAPS_PPT2','LDAPS_PPT3','LDAPS_PPT4']]
z=np.abs(zscore(f))
z
```

```python
print(np.where(z>3))
```

```python
#REMOVING OUTLIERS
ds11=ds1[(z<3).all(axis=1)]
ds11
```

Therefore the outliers have been removed

```python
#CHECKING FOR SKEWNESS
ds11.skew()
```

```python
#REMOVING SKEWNESS USING CUBEROOT METHOD
ds11[['LDAPS_PPT1','LDAPS_PPT2','LDAPS_PPT3','LDAPS_PPT4']] = np.cbrt(ds11[['LDAPS_PPT1','LDAPS_PPT2','LDAPS_PPT3','LDAPS_PPT4']])
ds11.skew()
```
